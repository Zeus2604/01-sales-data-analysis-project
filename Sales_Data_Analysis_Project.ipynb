# Sales Data Analysis Project

This project analyzes restaurant sales data to identify revenue patterns, customer behavior, and key business insights using Python, Pandas, and Matplotlib.

## Dataset
The dataset used in this project contains restaurant transaction records including total bill amount, customer gender, day of visit, time, and table size.

## Objectives

- Identify total revenue generated
- Determine which day generates the highest income
- Analyze revenue by customer gender
- Evaluate the distribution of bill amounts

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()
```

```python
df['total_bill'].sum()
```

```python
df['total_bill'].mean()
```

```python
df.groupby('day')['total_bill'].sum()
```

```python
df.groupby('sex')['total_bill'].sum()
```

```python
df.groupby('day')['total_bill'].sum().plot(kind='bar')
plt.title('Ingresos por día')
plt.show()
```

```python
df.groupby('sex')['total_bill'].sum().plot(kind='bar')
plt.title('Ingresos por género')
plt.show()
```

```python
df['total_bill'].plot(kind='hist')
plt.title('Distribución de montos de cuentas')
plt.show()
```

## Conclusiones

1. El sábado es el día con mayores ingresos.
2. Los clientes masculinos generan más ingresos que los femeninos.
3. El ticket promedio es aproximadamente 19.79.
4. La mayoría de los consumos se concentran entre 10 y 20.
